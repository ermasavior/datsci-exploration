## Read dataset "purchases.csv"
1. Download dataset "purchases.csv" in the course portal "IF4041 Ilmu Data dan Penggalian Data".
2. Put the dataset in the same folder with this jupyter file code.
3. Run this code below to read the csv file using pandas.

```python
import pandas as pd # import pandas library

# setting maximal rows and columns that can be shown
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 100)

# read csv file into 'df' dataframe
df = pd.read_csv('purchases.csv', sep = "\t")

# print row 1 untill 5 (with all columns)
print(df.iloc[0:5,:])
```

## Question 01 (Q01)

2. Write your own code in the block below, and show the result for the first ten rows.

```python
# write you own code for Q01 here
print(df.sort_values(by='Kota').iloc[0:10, :])
```

## Question 02 (Q02)
1. Sum the `Total` column  by `Kota`. <br>Hint: The are about 1 million data rows in `purchases.csv` and many of them have same city (`Kota`). Thus, you are asked to group all the data rows by `Kota`, and sum all the `Total` values. 
2. Write your own code in the block below, and show **all the result** (Kota - Total).

```python
# write you own code for Q02 here
sum_per_city = df.groupby(['Kota']).sum()
print(sum_per_city)
```

## Question 03 (Q03)
1. Regarding the result in Q02, write a code to show `Kota` and `Total` with the **higest value** in `Total`.
2. Regarding the result in Q02, write a code to show `Kota` and `Total` with the **lowest value** in `Total`.
3. Regarding the result in Q02, write a code to show the average value of `Total` (for all cities).

**HINTS:** The result data in Q02 can be in `dataframe`, `series` or other types, correspoding to the method you used to solve Q02. So, to calculate min, max and average of Q02 result, you need to be careful which API (`pandas dataframe`, `series`, etc) you need to use. Otherwise, you will get error.

```python
# write you own code for Q03 here
print('Max: ', sum_per_city.idxmax(), sum_per_city.max())

print('Min: ', sum_per_city.idxmin(), sum_per_city.min())

print('Avg: ', sum_per_city.mean())
```
